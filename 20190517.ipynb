```python
from sklearn.cluster import AgglomerativeClustering
from sklearn import linear_model
ncluster = 5
data11 = pd.read_csv('apts_68station.csv')
hc = AgglomerativeClustering(n_clusters=ncluster, affinity='euclidean', linkage='ward')  
hc.fit_predict(data11.T) 

#print(hc.labels_)

for i in range(ncluster):
    plt.plot(data11.iloc[:,hc.labels_==i].mean(axis=1),label=i)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
nor_lts = preprocessing.scale(proj,axis=0)
nor_pv = preprocessing.scale(projpv,axis=0)
nor_th850 = preprocessing.scale(projth850,axis=0)
nor_h500 = preprocessing.scale(projh500,axis=0)
nor_th700 = preprocessing.scale(projth700,axis=0)
nor_th925 = preprocessing.scale(projth925,axis=0)
nor_div850 = preprocessing.scale(projdiv850,axis=0)
nor_div925 = preprocessing.scale(projdiv925,axis=0)
nor_div700 = preprocessing.scale(projdiv700,axis=0)
#print(nor_lts)
plt.plot(nor_div850[:365,0], 'b-')
#plt.plot(projth700[:365,1], 'k-')
#plt.show()
```

```python
Y = (hc.labels_== 0)*1

Y = np.array(Y).reshape(len(Y),1)
pc=0
akk=np.empty((3642,180))
#akk[:,0:30] = nor_lts[0:3642,1:31]
#akk[:,30:60] = nor_pv[0:3642,1:31]
#akk[:,60:90] = nor_th850[0:3642,0:30]
#akk[:,90:120] = nor_h500[0:3642,1:31]
#akk[:,120:150] = nor_th700[0:3642,1:31]
akk[:,0:20] = nor_lts[0:3642,0:20]
akk[:,20:40] = nor_pv[0:3642,5:25]
akk[:,40:60] = nor_th925[0:3642,1:21]
akk[:,60:80] = nor_th850[0:3642,0:20]
akk[:,80:100] = nor_th700[0:3642,0:20]
akk[:,100:120] = nor_div925[0:3642,0:20]
akk[:,120:140] = nor_div850[0:3642,0:20]
akk[:,140:160] = nor_div700[0:3642,0:20]
akk[:,160:180] = nor_h500[0:3642,1:21]

aaa=pd.DataFrame(akk)

Xts = akk
print('Dimension of X: '+ str(Xts.shape))
print('Dimension of Y: '+ str(Y.shape))


mvreg = linear_model.LogisticRegression(C=1e6, solver='lbfgs', max_iter=500)
mvreg.fit(Xts, Y.flatten())

# Show the results
yhat_mv = mvreg.predict(Xts)

#print((mvreg.coef_, mvreg.intercept_))

# Show confusion matrix
ctl = pd.DataFrame({'y':Y.flatten(), 'yhat':yhat_mv.flatten()})
#print(ctl)
print('=== Confusion Matrix ===')
print(pd.crosstab(ctl['y'], ctl['yhat']))
from sklearn.metrics import accuracy_score, classification_report
cr = classification_report(ctl['y'], ctl['yhat'])
print(cr)
```

```python
import statsmodels.api as sm
mvlogit = sm.Logit(Y.flatten(), sm.add_constant(Xts))
mvlogit.fit(method='lbfgs', maxiter=500).summary()
```
